## **2. При необходимости — приведите такие значения как заработная плата к числовому типу.**

**Рассмотрим колонку `salary` - зарплата. Видно, что больше половины значений (40615 из 76028) - не пустые. Изучим структуру записей этой колонки подробнее:**

```python
ds['salary']
```

```python
# рассмотрим, например, 4-ю, 76380-ю и 76381-ю записи:
ds['salary'][4], ds['salary'][76380], ds['salary'][76381]
```

**Видим, что структура записи соответствует питоновскому словарю с 4-мя ключами и напоминает формат json. Однако для формата json характерны двойные кавычки, а также написание логических переменных с маленькой буквы. Поправив эти нюансы, можно легко преобразовать строчку зарплаты в питоновский словарь:**

```python
import json

# функция преобразования строчного формата записи salary в питоновский словарь. Возвращает словарь
def get_salary_dict(salary):
    salary = salary.replace("'", '"')
    salary = salary.replace('True', 'true')
    salary = salary.replace('False', 'false')
    salary = salary.replace('None', 'false')
    return json.loads(salary)

# проверим работу функции, выведем получившийся словарь, например, для 4-й записи колонки salary
get_salary_dict(ds['salary'][4])
```

**Кроме того, видим также, что есть несколько схем отображения зарплаты. В основном указывается зарплатная вилка (от минимальной до максимальной суммы), но есть и такие схемы, где указываются либо только минимум, либо только максимум. Также схемы различаются по валюте зарплаты и по тому, указана ли сумма до вычета налогов(gross), или после (net). Посчитаем, сколько таких возможных схем есть в датасете и можем ли какими-то из них пренебречь:**

```python
from numpy import NaN

# функция возвращает структуру записи зарплаты (схему)
def get_salary_scheme(salary):
    if salary is NaN:
        return NaN
    else:
        salary_dct = get_salary_dict(salary)
        scheme = ('Min' if salary_dct['from'] else 'None') + '-'
        scheme += ('Max' if salary_dct['to'] else 'None') + ' ' + salary_dct['currency'] + ' '
        scheme += 'gross' if salary_dct['gross'] else 'net'
        return scheme
```

```python
# создадим пандас-серию 'salary_scheme'
salary_scheme = ds['salary'].apply(get_salary_scheme)
salary_scheme
```

```python
# выведем получившиеся структуры записи зарплаты (схемы) и их количество
schemes = salary_scheme.value_counts()
schemes
```

**Видим, что основная часть данных по зарплате представлена в рублях (первые 6 строчек получившегося списка). Посчитаем, какую долю от общего количества зарплатных записей занимают зарплаты в иностранной валюте:**

```python
# посчитаем долю 'зарплат в иностранной валюте' в общем количестве записей
print(f'all: {schemes.sum()}     RUR: {schemes[:6].sum()}     other currency: {schemes[6:].sum()}    ',
      f'other/all: {round(schemes[6:].sum()*100/schemes.sum(), 2)}%')
```

**Видим, что записи, где зарплата указана в иностранной валюте составляют менее полутора процентов от всех записей с указанием зарплаты. Можем пренебречь ими в дальнейшем, чтобы не зависеть в нашем исследовании от курсов валют**

**Cоздадим теперь 2 колонки: с минимальной и максимальной зарплатой (`'min_salary'` и `'max_salary'`, соответственно) В новых колонках приведём все зарплаты к значениям "суммы на руки", (т.е. после вычета налогов): если зарплата указана gross (до вычета налогов), то вычтем из нее 13% НДФЛ**

```python
# функция возвращает минимальную или максимальную зарплату, в зависимости от параметра from_to
def get_salary(salary, from_to):
    if salary is NaN:
        return NaN
    else:
        salary_dct = get_salary_dict(salary)
        if salary_dct['currency'] != 'RUR' or not salary_dct[from_to]:
            return NaN
        result = float(salary_dct[from_to])
        if salary_dct['gross']:
            result = result * 0.87
        return result

# функция возвращает минимальную зарплату (параметр from_to = 'from')
def get_min_salary(salary):
    return get_salary(salary, 'from')

# функция возвращает максимальную зарплату (параметр from_to = 'to')
def get_max_salary(salary):
    return get_salary(salary, 'to')
```

```python
ds['min_salary'] = ds['salary'].apply(get_min_salary)
ds['max_salary'] = ds['salary'].apply(get_max_salary)
ds.info()
```

**Видим, что каждая запись имеет структуру в виде списка словарей с единственным ключом `name`, по которому и лежат наименования ключевых навыков. Проще всего на основе такой записи создать список ключевых навыков:**

```python
# разберем, к примеру строку с ключевыми навыками из 1-й записи
skill_string = ds['key_skills'][1]

# создадим список на основе исходной строки
skill_list = skill_string[1:-1].replace("{'name': '", '').replace("'}", '').split(', ')
skill_list
```

**Пройдемся циклом по всем записям и заполним словарь `skills_dct`, где ключом будет наименование навыка, а значением - количество раз, которое этот навык встречается в датасете:**

```python
skills_dct = dict()

for skill_string in ds['key_skills']:
    if skill_string != '[]':
        # создадим список на основе текущей строки
        skill_list = skill_string[1:-1].replace("{'name': '", '').replace("'}", '').split(', ')
        for skill in skill_list:
            # заполнение словаря
            skills_dct[skill] = skills_dct.get(skill, 0) + 1

# выведем 25 значений словаря
list(skills_dct.items())[:25]
```

```python
print(f"В словаре 'skills_dct' {len(skills_dct)} ключевых навыков")
```

**На основе этого словаря создадим пандас-серию и отсортируем по убыванию значений (количества упоминаний ключевых навыков в вакансиях)**

```python
skills = pd.Series(skills_dct).sort_values(ascending=False)
skills
```

```python
print(f"Все ключевые навыки упоминаются в датасете в общей сложности {skills.sum()} раз")
```

#### **5.3. Анализ колонки `specializations`**

**Рассмотрим подробнее структуру записей в колонке `specializations` (сферы деятельности), выведем 0-ю, 10-ю, 30-ю записи:**

```python
ds['specializations']
```

```python
# записи 0, 10, 30
ds['specializations'][0], ds['specializations'][10], ds['specializations'][30]
```

**Видим, что каждая запись имеет структуру в виде списка словарей, в каждом из которых по ключу `profarea_name` лежит название специализации (сферы деятельности). Почистим записи, оставим только перечисленные через точку с запятой сферы деятельности. Выведем для примера первые 35 записей:**

```python
def get_profarea(specializations):
    # разделяем строку по нужному ключу, нулевой элемент списка выкидываем
    spec_list = specializations.split("'profarea_name': '")[1:]
    # удаляем все лишнее из каждого элемента списка
    spec_list = [spec.split("'")[0] for spec in spec_list]
    # удаляем дубликаты, сортируем, объединяем список в строку и возвращаем строку
    return '; '.join(sorted(list(set(spec_list))))

# выведем для примера первые 35 почищенных записей со специализациями (сферами деятельности)
for i in range(35):
    spec_string = ds['specializations'][i]
    print(get_profarea(spec_string))
```

**Добавим колонку со специализацией (сферой деятельности) '`profarea`' в датасеты `ds` и `ds2`**

```python
ds['profarea'] = ds['specializations'].apply(get_profarea)
ds2['profarea'] = ds2['specializations'].apply(get_profarea)
ds.info()
print()
ds2.info()
```

```python
ds['profarea'].value_counts()[:25]
```

**Видим, что очень много сфер деятельности связано с Информационными технологиями, интернетом, телекомом. Объединим и назовем для краткости эту сферу IT, обновим колонку `profarea` в датасетах**

```python
def get_it(profarea):
    return 'IT' if 'Информационные технологии' in profarea else profarea

ds['profarea'] = ds['profarea'].apply(get_it)
ds2['profarea'] = ds2['profarea'].apply(get_it)
ds['profarea']
```
